```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna('MS', inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preo'].mean()
```

```python
vendas['Preo'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
mediana_preco = vendas['Preo'].median()
mediana_preco
```

```python
vendas.head()
```

```python
vendas[['Preo','Preo_com_frete']].describe()
```

```python
preco_errado = vendas.query('Preo > Preo_com_frete')
vendas_correto = vendas.query('Preo < Preo_com_frete')
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado =='SP'").describe()
```

```python
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_correto.groupby('Nome_Departamento')['Preo_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```python
vendas_correto['mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head()
```

```python
vendas_correto.head()
```

```python
cliente[['idade','renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda' : 'float'})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how='left', on='cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='green')
plt.ylabel('Mdia de Idade')
plt.title('Idade Mdia por Bandeira', loc='center',fontsize=18, fontweight='bold',color='black');
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='blue')
plt.ylabel('Mdia de Renda')
plt.title('Mdia de Renda por Canal de Venda',loc='center',fontsize=18, fontweight='bold',color='black');
```

```python
vendas_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(20,12))
plt.xlabel('Data da Venda')
plt.ylabel('Quantidade de Vendas')
x = vendas_por_data['Data']
y = vendas_por_data['idcompra']
plt.plot(x,y,color='green',linewidth=2)
plt.show()
```

```python
agg_depto_preco = round(vendas_correto.groupby('Nome_Departamento')['Preo_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_depto_preco
```

```python
fig = plt.figure(figsize=(30,12))
plt.bar(agg_depto_preco['Nome_Departamento'], agg_depto_preco['Preo_com_frete'], color='blue')
plt.ylabel('Mdia de Preo')
plt.title('Mdia de Preo por Departamento',loc='center',fontsize=18, fontweight='bold',color='black');
```
